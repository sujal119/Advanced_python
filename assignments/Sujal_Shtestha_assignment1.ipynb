## Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.


````python
marks = [45, -3, 88, "NA", 102, None, 67]

def remove_invalid_marks(marks):
    valid_marks = []
    invalid_count = 0
    for mark in marks:
        if isinstance(mark, int) and 0 <= mark <= 100:
            valid_marks.append(mark)
        else:
            invalid_count += 1
    valid_percentage = (len(valid_marks) / len(marks)) * 100
    return valid_marks, invalid_count, valid_percentage
print(remove_invalid_marks(marks))
````

## Comments:
From the list, we remove invalid entities and return valid marks, no of invalid_marks and valid_marks in percentage using fucntion.

## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.


````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print("Prime numbers between 1 and 50:")
for num in range(1, 51):
    if is_prime(num):
        print(num, end=" ")
````

## Comments:
In this question, we create function named "is_prime" to display prime numbers between 1 to 50.

## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python

def parse_student(data):
    try:
        name, roll, age, marks = data.split("|")
        roll = int(roll)
        age = int(age)
        marks = [int(m.strip()) for m in marks.split(",")]

        if age < 16 or any(m < 0 or m > 100 for m in marks):
            return {}

        return {roll: {"name": name, "age": age, "marks": marks}}

    except:
        return {}

print(parse_student("Ram|101|22|78, 88, 92"))
````

## Comments:
Here, we create a function that accepts a raw input string in the format asked in question above and extract name, roll number, age and marks & convert them into their appropriate type and hence obtain result as a dictionary using roll number as the key.

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
def separate_data(data):
    numeric = []
    non_numeric = []

    for d in data:
        if isinstance(d, (int, float)) and not isinstance(d, bool):
            numeric.append(float(d))
        else:
            non_numeric.append(d)

    return numeric, non_numeric

data = [10, "20", 30.5, "abc", None, True]
print(separate_data(data))
````

## Comments:
Here, we create a function to seperate numeric and non-numeric values and convert numeric values to float. 
True behaves like 1 because bool is a subclass of int in Python

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
def analyze(records):
    averages = {k: sum(v)/len(v) for k, v in records.items()}
    top = max(averages, key=averages.get)
    bottom = min(averages, key=averages.get)

    return {
        "averages": averages,
        "topper": top,
        "lowest": bottom
    }

records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

print(analyze(records))

    
````

## Comments: 
Here, we create function that computes average per student and compare each student average to identify top and bottom performers and return a result in single dictionary.


## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
def analyze_school(data):
    student_avg = {}
    class_toppers = {}
    total_marks = 0
    total_count = 0

    for cls, students in data.items():
        best_student = None
        best_avg = 0

        for name, marks in students.items():
            avg = sum(marks) / len(marks)
            student_avg[name] = avg
            total_marks += sum(marks)
            total_count += len(marks)

            if avg > best_avg:
                best_avg = avg
                best_student = name

        class_toppers[cls] = (best_student, best_avg)

    school_avg = total_marks / total_count

    return {
        "class_toppers": class_toppers,
        "school_average": school_avg,
        "student_averages": student_avg
    }
data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
analyze_school(data)
````

## Comments:
Here, we create a function that calculates average of the students of both class A and B to find topper in each class and also calculate student_average value.
Nested dictionaries are ideal for hierarchical data like school → class → students as they organize information in multiple levels, mimicking real-world structures.
